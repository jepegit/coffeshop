## Some matplotlib examples

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Create a simple plot

```python
# Create some data to plot
x = np.linspace(-5, 5, 50)
```

```python
# Plot the data
plt.figure()
plt.plot(x, 5*x, 'r--', x, x**2, 'b.', x, 5*x**2, 'g^')  # Plot 5x with red dashes, x^2 with blue dots and 5x^2 with green triangles
# Adding x and y labels
plt.xlabel('x')
plt.ylabel('y')  #
plt.title('Simple plot')
plt.legend(['2x', r'x$^2$', r'2x$^2$'])
plt.show()
```

You can write TeX equations in labels, legend and title by adding $ around the TeX expression.
